**Unsupervised Clustering Algorithms**

# KMeans

```python
# Import necessary libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Generate Synthetic
X,y = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
#Apply K-Means clustering
kmeans= KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
#get cluster lables and centroids
kmeans_labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolors='k', label='Ground Truth')
```

```python
# Plot the cluster centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
```

```python
# Plot the data points with colors representing the predicted clusters by K-means
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', marker='s', edgecolors='k', alpha=0.5, label='K-means Predictions')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

# Agglomerative Clustering

```python
# Import necessary libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data using make_blobs
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
# Apply Agglomerative Hierarchical Clustering
agg_clustering = AgglomerativeClustering(n_clusters=4)
agg_labels = agg_clustering.fit_predict(X)
```

```python
# Plot the original data points with colors representing the ground truth clusters
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolors='k', label='Ground Truth')
```

```python
# Plot the data points with colors representing the predicted clusters by Agglomerative Clustering
plt.scatter(X[:, 0], X[:, 1], c=agg_labels, cmap='viridis', marker='s', edgecolors='k', alpha=0.5, label='Agglomerative Cluster')
plt.title('Agglomerative Hierarchical Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

# DBSCAN

```python
# Import necessary libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data using make_blobs
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
# Apply DBSCAN clustering
dbscan_clustering = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels = dbscan_clustering.fit_predict(X)
```

```python
# Plot the original data points with colors representing the ground truth clusters
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolors='k', label='Ground Truth')
```

```python
# Plot the data points with colors representing the predicted clusters by DBSCAN
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', marker='s', edgecolors='k', alpha=0.5, label='DBSCAN Predictetions')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

# Hierarchical Clustering

```python
# Import necessary libraries
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data using make_blobs
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
# Apply Hierarchical Clustering
linkage_matrix = linkage(X, method='ward')
```

```python
# Plot the dendrogram
plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
